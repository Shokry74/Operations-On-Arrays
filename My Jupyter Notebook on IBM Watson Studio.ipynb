### Next line will be about do Mean Square Error 
MSE = (1/n) * np.sum(np.square(predictions - labels))

```python
import numpy as np 
a = np.array([[1,2,3,4]])
b = np.array([[5,2,5,4]])
n = 3
x = a - b
z = (1/n) * np.sum(np.square(x))
print("Mean Square Error = " + str(z))
```
